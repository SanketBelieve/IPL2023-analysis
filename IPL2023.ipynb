```python
df=pd.read_csv('each_match_records.csv')
```

```python
plt.figure(figsize=(20,10))
sns.countplot(x='winner',data=df)
plt.title('No of matches win by Teams')
plt.xticks(rotation=90, ha="right")
plt.show()
```

```python
plt.figure(figsize=(8,8))
#plt.figure(figsize=(8, 8))
df['toss_decision'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Toss Decision')
plt.ylabel('')
plt.show()
```

```python
plt.figure(figsize=(20,10))
sns.countplot(x='man_of_match',data=df)
plt.title('Man of the match')
plt.xticks(rotation = 90,ha = 'right')
plt.show()
```

```python
plt.figure(figsize=(10,10))
sns.countplot(x='match_referee',data=df)
plt.title('No of Matches Host as Refree')
plt.xticks(rotation = 90,ha = 'right')
plt.show()
```

```python
first_bat_win = df['winner_runs'].notna().sum()

# Display the result
print(f"Number of non-empty entries in the column: {first_bat_win}")

second_bat_wins = df['winner_wickets'].notna().sum()
print(f"Number of non-empty entries in the column: {second_bat_wins}")
```

```python
# Create a pie chart
labels = ['First Bat Wins', 'Second Bat Wins']
values = [first_bat_win, second_bat_wins]

plt.pie(values, labels=labels, autopct='%1.1f%%', startangle=90)
plt.axis('equal')
plt.title('N')

# Display the plot
plt.show()
```
